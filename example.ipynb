```python
import nltk
sentence = "I wakeup at 6:00 AM and breakefast at 7:30 AM and take bath after that going to the bus at 8:15 AM"
from nltk.tokenize import word_tokenize
nltk.download("punkt_tab")
tokens = word_tokenize(sentence)
print(tokens)
```
